```python
from gensim.test.utils import common_texts, get_tmpfile
from gensim.models import Word2Vec
import gensim
import logging

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

**Memuat data wikipedia**

Masukan path dari data wikipedia yang sudah di download, lalu masukkan kedalam variabel data_file

```python
data_file = "./wiki_id.txt"

with open(data_file, 'rb') as f:
    for i,line in enumerate(f):
        print(line)
        break
```

**Membaca input text wikipedia dan dimasukkan kedalam variabel documents dalam bentuk list**

```python
def read_input(input_file):
    logging.info("reading file, this may take a while".format(input_file))
    
    with open(input_file, 'rb') as f:
        for line in f:
            yield gensim.utils.simple_preprocess(line)
            

documents = list(read_input(data_file))
print(documents)
logging.info("Done reading data file")
```

**Membuat model dan training word2vec dari data text wikipedia**

```python
model = gensim.models.Word2Vec(documents, size=100, window=5, min_count=2, workers=3)
model.train(documents, total_examples=len(documents), epochs=10)
```

**Menyimpan model word2vec kedalam bentuk .txt**

```python
filename = "./wiki_word2vec.txt"
model.wv.save_word2vec_format(filename, binary = False)
```
